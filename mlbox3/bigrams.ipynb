```python
from collections import Counter
import pandas
documents = {
    'l1': "it was the best of times",
    'l2': "it was the worst of times",
    'l3': "it was the age of wisdom",
    'l4': "it was the age of foolishness"
}

all_words = ' '.join([documents[key] for key in documents.keys()])
documents['all'] = all_words
```

```python
bigrams = set([b for l in documents.values() for b in zip(l.split(" ")[:-1], l.split(" ")[1:])])
bigrams = [' '.join(x) for x in bigrams]
```

```python
l1_multi_hot = [1 if w in documents['l1'] else 0 for w in bigrams]
l2_multi_hot = [1 if w in documents['l2'] else 0 for w in bigrams]
l3_multi_hot = [1 if w in documents['l3'] else 0 for w in bigrams]
l4_multi_hot = [1 if w in documents['l4'] else 0 for w in bigrams]
all_multi_hot = [1 if w in documents['all'] else 0 for w in bigrams]
```

```python
X = [l1_multi_hot, l2_multi_hot, l3_multi_hot, l4_multi_hot, all_multi_hot]
df = pandas.DataFrame(X, columns=list(bigrams))
df.T.rename(columns={0:'l1',1:'l2',2:'l3',3:'l4',4:'all'})
```
